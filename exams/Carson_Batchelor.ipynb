### \#4: Write a program which repeatedly solicits strings (words) from the user until the user enters the string “done”. Once “done” is entered, print out the total number of words entered (without regard to the language of origin or whether the string is an actual word), total number of characters entered, length of the longest string entered, length of the shortest string entered, and the average number of characters in each string (word). If the user enters anything other than an alphabetical character, detect their mistake using try and except and print an error message while skipping to the next string (8 points). 

```python
total_words = 0
characters_per_word = []

while True:
    try:
        # Get input from the user
        user_input = input("Enter a word (or 'done' to finish): ")
        # Check if done
        if user_input.lower() == 'done':
            break
        # Check if its only letters
        if not user_input.isalpha():
            raise ValueError("Only alphabetical characters allowed")
        # Update values
        total_words += 1
        characters_per_word.append(len(user_input)) 
        longest_word = max(characters_per_word)
        shortest_word = min(characters_per_word)
    
    except ValueError as e:
        print(e)
        continue

total_characters = sum(characters_per_word)

if total_words != 0:
    avg_length = total_characters/total_words
else:
    avg_length = None

print(f"Total words: {total_words}")
print(f"Total characters: {total_characters}")
print(f"longest word: {longest_word}")
print(f"Shortest word: {shortest_word}")
print(f"Average number of characters in all words: {avg_length}")
```
