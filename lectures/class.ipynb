```python
import pandas as pd
from pandas_datareader import data
import numpy as np

gdp_data = data.DataReader("GDPC1", "fred", "1962-01-01", "2017-07-01")
gdp_data
```

```python
gdp_data["Log GDP"] = np.log(gdp_data["GDPC1"])

gdp_data["GDP Growth Rate"] = gdp_data["Log GDP"].diff() * 400

gdp_data
```

```python
gdp_data["Log GDP Growth Rate"] = gdp_data["GDP Growth Rate"].shift(1)

gdp_data
```

```python
import statsmodels as sm

gdp_data2 = gdp_data.dropna()

model = sm.tsa.autoreg(gdp_data2["GDP Grow Rate"], 1)
```
