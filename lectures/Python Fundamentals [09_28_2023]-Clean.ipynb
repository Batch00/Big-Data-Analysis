# Measuring Central Tendency & Spread

The most common measure of central tendency is the mean. For skewed distribution or when there is concern about outliers, the median may be preferred. It is seldom advantageous to consider the mode.

The variance and standard deviation measure spread. Variance is the mean of the squares of the individual deviations. 
Standard deviation is the square root of the variance. For normally distributed data, approximately 95% of the values lie within two standard deviations of the mean. A third measure of spread is the interquartile range. Quartiles of a distribution are three values which divide it into fourths. We define the IQR as IQR = Q3 − Q1. This is the middle half of the data
The interquartile range (IQR) is a robust measure of spread.

Describing the data numerically can be a helpful, but sometimes laborious and inefficient approach to identifying patterns. A common purpose of univariate data exploration is to get a 'feeling' for the values a particular variable is likely or unlikely to take on. At the same time, we are likely going to be interested in the most extreme values it has assumed as a proxy for upper and lower bounds, as well as measures describing the range of values that occur most frequently in the data. Boxplots achieve all of these ends in a simple, easy to read (and compare) graphic. Here is an example of a boxplot.

A boxplot is a graph that gives you a good indication of how the values in the data are spread out.

What is also noteable is the dispersion in the data. The whiskers extending above and below the box extend 1.5 times the IQR (25th - 75th percentiles). Since the IQR for batting average is .280-.240 = .040, the whiskers are +- .060, terminating at about .180 and .340. Any points beyond these whiskers are typically considered to be 'outliers'. Another useful way to visualize data is to generate a histogram. Histograms plot counts or proportions of data in disjoint intervals along the support set.  Actual values of the data are not used; a datum’s presence in a particular interval (or “bin”) is merely tallied. The histogram then generates vertical bars, the heights of which represent the tally or the proportion of the data falling in each bin. Generating a historgram is very useful in determining if the distribution of values is symmetric or skewed, whether there is more than one mode, and whether outliers appear in the data. The choice of how many bins to use can greatly impact the visual representation of a histogram. 

Matplotlib's `ax.hist()` method can be used to build a histogram. The first argument passed to `ax.hist()` corresponds to the ldata you wish to plot. The second argument corresponds to the number of bins in which to place the data, i.e. the number of bars on the histogram. 


```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()


ax.hist(bb['Batting average'], bins=[.1,.2,.3,.4,.5,.6,.7,.8,.9,1])
ax.set_title('Example Historgram')
ax.set_xlabel('Batting Average')
ax.set_ylabel('Number of Players')

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()


ax.hist(bb['Batting average'], 10, color='red', edgecolor='blue', alpha=.5)
ax.set_title('Example Historgram')
ax.set_xlabel('Batting Average')
ax.set_ylabel('Number of Players')


fig.tight_layout()
```

It appears that the distribution of batting average among players active during the 1991 and 1992 MLB seasons was uni-modal and roughtly symmetrical. In addition to picking the number of bins, there are several other arguments that can be passed to customize the display. These include specific bin ranges, axis labels, colors, titles, and more. JSON is a common data format and the name stands for JavaScript Object Notation. It is often used when data is sent from a server to a web page, as well as for API queiries. That is because JSON is "self-describing" and easy to understand: all data comes in name/value pairs and is separated by commas. That's kinda ugly and hard to read. How can we make this look a little prettier? If you can't read the data very well it will be hard to analyze it. 

One option is to use the `json` module.The `json` module makes it easy to parse JSON strings and files containing JSON object. To write JSON to a file in Python, we can use `json.dump()` method.

```python
import json

astronauts = requests.get('http://api.open-notify.org/astros.json')
data = astronauts.json()
with open('astronauts.json', 'w') as f:
    json.dump(data, f)
    
data
```

Above we saved our json file to our current working directory. You can use `json.loads()` method to read a file containing JSON object. We use the `open()` function to read the json file. Then, the file is parsed using `json.loads()` method which gives us a dictionary named samedata. 

```python
with open('astronauts.json','r') as f:
    samedata = json.loads(f.read())
```

```python
type(samedata)
```

The value of 'people' is multiple levels (known as a nested dict). By calling` pd.json_normalize(json_obj)`, we get:

```python
df = pd.json_normalize(samedata, record_path =['people'])
df
```

The results look much better. All nested values are flattened and converted into separate columns.
