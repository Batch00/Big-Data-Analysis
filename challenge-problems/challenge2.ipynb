```python
%matplotlib inline
from pandas_datareader import data, wb
import matplotlib.pyplot as plt
import datetime as dt

code = "CPILFESL"
start = dt.datetime(1960,1,1)
fred = data.DataReader(code, 'fred', start)
fred['Inflation Rate'] = fred['CPILFESL'].pct_change() * 100

# Calculate historical average inflation rate
historical_avg_inflation_rate = fred['Inflation Rate'].mean()

# Plot inflation Rate
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(fred.index, fred["Inflation Rate"], color="blue", label="Inflation Rate")
ax.axhline(y=historical_avg_inflation_rate, color='r', linestyle='--', label='Historical Average')
ax.set_xlabel('Year')
ax.set_ylabel('Inflation Rate (%)')
ax.set_title('Consumer Price Index for All Urban Consumers: All Items Less Food and Energy (Inflation Rate)')
ax.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
